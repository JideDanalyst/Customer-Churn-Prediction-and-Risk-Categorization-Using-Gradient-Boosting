```python
cust['Email_Opt_In'] = cust['Email_Opt_In'].astype(int)
cust['Target_Churn'] = cust['Target_Churn'].astype(int)

numeric_columns = [
    'Age', 'Annual_Income', 'Total_Spend', 'Years_as_Customer',
    'Num_of_Purchases', 'Average_Transaction_Amount', 'Num_of_Returns',
    'Num_of_Support_Contacts', 'Satisfaction_Score', 'Last_Purchase_Days_Ago',
    'Email_Opt_In'
]

X = cust[numeric_columns]
y = cust['Target_Churn']

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5]
}

gb_model = GradientBoostingClassifier(random_state=42)

grid_search = GridSearchCV(estimator=gb_model, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)

grid_search.fit(X_poly, y)

best_params = grid_search.best_params_
print("Best parameters found: ", best_params)

gb_model_tuned = GradientBoostingClassifier(
    n_estimators=best_params['n_estimators'],
    learning_rate=best_params['learning_rate'],
    max_depth=best_params['max_depth'],
    random_state=42
)

cv_scores = cross_val_score(gb_model_tuned, X_poly, y, cv=5)
print("Cross-validation scores: ", cv_scores)
print("Mean cross-validation score: ", cv_scores.mean())

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X_poly, y)

gb_model_tuned.fit(X_resampled, y_resampled)

y_pred = gb_model_tuned.predict(X_poly)

print("Accuracy:", accuracy_score(y, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
print("Classification Report:\n", classification_report(y, y_pred))
```

```python
X = cust[numeric_columns]
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Predict churn probabilities
y_scores = gb_model_tuned.predict_proba(X_poly)[:, 1]  # Probability of class 1 (churn)

y_pred = gb_model_tuned.predict(X_poly)

predictions_df = cust.copy()
predictions_df['Churn_Probability'] = y_scores
predictions_df['Churn_Prediction'] = y_pred

churned_customers = predictions_df[predictions_df['Churn_Prediction'] == 1]
not_churned_customers = predictions_df[predictions_df['Churn_Prediction'] == 0]
```

```python
def categorize_churn(prob):
    if prob > 0.75:
        return 'High Risk'
    elif prob > 0.50:
        return 'Medium Risk'
    else:
        return 'Low Risk'

predictions_df['Churn_Risk'] = predictions_df['Churn_Probability'].apply(categorize_churn)

high_risk_customers = predictions_df[predictions_df['Churn_Risk'] == 'High Risk']
medium_risk_customers = predictions_df[predictions_df['Churn_Risk'] == 'Medium Risk']
low_risk_customers = predictions_df[predictions_df['Churn_Risk'] == 'Low Risk']
```
